**Reading the table of July-2018, 2020, 2022 with selected columns**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

# Didn't merge any tables of different years 
air_2018_df = pd.read_csv('./07-2018 Airport.csv')
air_2020_df = pd.read_csv('./07-2020 Airport.csv')
air_2022_df = pd.read_csv('./07-2022 Airport.csv')

air_2018_df.head(10)
```

```python
air_2018_df.info()
```

**Exploring cancelled airlines or airports**

```python
# if variable 'cancelled' == 1, then it's cancelled. 
cancelled_2018 = air_2018_df[air_2018_df['CANCELLED']==1]
cancelled_2018
```

```python
'''
Variable 'CANCELLATION_CODE'
A: Carrier
B: Weather
C: National Air System
D: Security

cc_a = cancelled airline due to carrier (A)..
'''
cc_a_2018 = cancelled_2018[cancelled_2018['CANCELLATION_CODE']=='A']
cc_b_2018 = cancelled_2018[cancelled_2018['CANCELLATION_CODE']=='B']
cc_c_2018 = cancelled_2018[cancelled_2018['CANCELLATION_CODE']=='C']
cc_d_2018 = cancelled_2018[cancelled_2018['CANCELLATION_CODE']=='D']
```

```python
# counting occasions by the cancellation_code
print(len(cc_a_2018))
print(len(cc_b_2018))
print(len(cc_c_2018))
print(len(cc_d_2018))
```

```python
# 2020 cancellation
cancelled_2020 = air_2020_df[air_2020_df['CANCELLED']==1]

cc_a_2020 = cancelled_2020[cancelled_2020['CANCELLATION_CODE']=='A']
cc_b_2020 = cancelled_2020[cancelled_2020['CANCELLATION_CODE']=='B']
cc_c_2020 = cancelled_2020[cancelled_2020['CANCELLATION_CODE']=='C']
cc_d_2020 = cancelled_2020[cancelled_2020['CANCELLATION_CODE']=='D']

print(len(cc_a_2020))
print(len(cc_b_2020))
print(len(cc_c_2020))
print(len(cc_d_2020))
```

```python
# 2022 cancellation
cancelled_2022 = air_2022_df[air_2022_df['CANCELLED']==1]

cc_a_2022 = cancelled_2022[cancelled_2022['CANCELLATION_CODE']=='A']
cc_b_2022 = cancelled_2022[cancelled_2022['CANCELLATION_CODE']=='B']
cc_c_2022 = cancelled_2022[cancelled_2022['CANCELLATION_CODE']=='C']
cc_d_2022 = cancelled_2022[cancelled_2022['CANCELLATION_CODE']=='D']

print(len(cc_a_2022))
print(len(cc_b_2022))
print(len(cc_c_2022))
print(len(cc_d_2022))
```

**Exploring delays**

```python
'''
CarrierDelay
Carrier Delay, in Minutes

WeatherDelay
Weather Delay, in Minutes

NASDelay
National Air System Delay, in Minutes

SecurityDelay
Security Delay, in Minutes

LateAircraftDelay
Late Aircraft Delay, in Minutes

da_c = delayed airline due to carrier..
'''

da_c_2018 = air_2018_df[air_2018_df['CARRIER_DELAY']==1]
da_w_2018 = air_2018_df[air_2018_df['WEATHER_DELAY']==1]
da_n_2018 = air_2018_df[air_2018_df['NAS_DELAY']==1]
da_s_2018 = air_2018_df[air_2018_df['SECURITY_DELAY']==1]
da_l_2018 = air_2018_df[air_2018_df['LATE_AIRCRAFT_DELAY']==1]
```

```python
# counting occasions by the cause of delay
print(len(da_c_2018))
print(len(da_w_2018))
print(len(da_n_2018))
print(len(da_s_2018))
print(len(da_l_2018))
```

```python
# 2020 Delays
da_c_2020 = air_2020_df[air_2020_df['CARRIER_DELAY']==1]
da_w_2020 = air_2020_df[air_2020_df['WEATHER_DELAY']==1]
da_n_2020 = air_2020_df[air_2020_df['NAS_DELAY']==1]
da_s_2020 = air_2020_df[air_2020_df['SECURITY_DELAY']==1]
da_l_2020 = air_2020_df[air_2020_df['LATE_AIRCRAFT_DELAY']==1]
print(len(da_c_2020))
print(len(da_w_2020))
print(len(da_n_2020))
print(len(da_s_2020))
print(len(da_l_2020))
```
